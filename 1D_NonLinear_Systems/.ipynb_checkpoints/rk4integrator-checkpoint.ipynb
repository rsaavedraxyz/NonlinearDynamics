```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def NonLinearODE(t, x):
    # fun = 1.0 + 0.5 * np.cos(x)
    # fun = np.exp(x) - np.cos(x)
    # dxdt = 1.0 - 2.0 * np.cos(x)
    # dxdt = np.tan(x)
    # dxdt = 1 - np.exp(-x*x)
    # dxdt =  - x**3
    # dxdt = x**2 * (6 - x)
    dxdt = np.log(x)
    return dxdt
```

```python
initial_conditions = np.arange(0.2,0.8,0.1)

plt.figure(1, figsize=(5,5))

dt = 0.0001
for ic in initial_conditions:
    t = np.arange(0, 0.5, dt)
    x = np.zeros_like(t)
    x[0] = ic
    for i in range(len(t)-1):
        k1 = NonLinearODE(i, x[i]);
        k2 = NonLinearODE(i + 0.5*dt, x[i] + 0.5*dt*k1); 
        k3 = NonLinearODE(i + 0.5*dt, x[i] + 0.5*dt*k2);
        k4 = NonLinearODE(i + dt, x[i] + dt*k3);
        x[i + 1] = x[i] + (1/6)*dt*(k1 + 2*k2 + 2*k3 + k4) 
    plt.plot(t,x)

plt.xlabel("t", size=16)
plt.ylabel("x", size=16)
```
