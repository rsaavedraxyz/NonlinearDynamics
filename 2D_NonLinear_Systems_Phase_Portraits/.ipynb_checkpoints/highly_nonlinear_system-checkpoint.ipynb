# Highly non-linear system

$$
\left\lbrace
\begin{align}
\dot{x} &= y^3 - 4x \\
\dot{y} &= y^3 - y -3x
\end{align}
\right..
$$

## Numerical analysis

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_eigen(_A):
    """
    Get the eigenvalues and eigenvectors of the matrix _A
    """
    eigval, eigvec = np.linalg.eig(_A(0, 0))
    l1 = eigval[0]
    l2 = eigval[1]
    v1=np.array(np.transpose(eigvec[:,0]))
    v2=np.array(np.transpose(eigvec[:,1]))
    return l1,l2,v1,v2

def plot_vector_field(_lims, _vfield):
    """
    Plot the vector field _vfield in a grid with limits _lims
    """
    xlims = _lims[0]
    ylims = _lims[1]
    xlo,xhi,ylo,yhi = xlims[0],xlims[1],ylims[0],ylims[1]
    XX, YY = np.meshgrid(np.arange(xlo,xhi,0.01), np.arange(ylo,yhi,0.01))
    DX, DY = _vfield([XX, YY], 0)
    plt.streamplot(XX,YY,DX,DY,color='darkcyan',density=2.,minlength=0.2,linewidth=1,arrowsize=1.5)
    
def rk4(dxdt, x, t, dt, *args, **kwargs):
    x = np.asarray(x)
    k1 = np.asarray(dxdt(x, t, *args, **kwargs))*dt
    k2 = np.asarray(dxdt(x + k1*0.5, t, *args, **kwargs))*dt
    k3 = np.asarray(dxdt(x + k2*0.5, t, *args, **kwargs))*dt
    k4 = np.asarray(dxdt(x + k3, t, *args, **kwargs))*dt
    return x + (k1 + 2*k2 + 2*k3 + k4)/6
```

```python
def V(z, t):
    x = z[0]
    y = z[1]
    dxdt = y**3 - 4*x
    dydt = y**3 - y - 3*x
    return [dxdt, dydt]
```

```python
plt.figure(figsize=(8,8))

# Set boundaries
scl = 20.
xlo,xhi,ylo,yhi = -1.*scl, 1.*scl, -1.*scl, 1.*scl
lims = [[xlo,xhi],[ylo,yhi]]

# Fixed points
plt.scatter(0.0,0.0,fc='w',ec='r',s=80)
plt.scatter(-2.0,-2.0,fc='w',ec='r',s=80)
plt.scatter(2.0,2.0,fc='w',ec='r',s=80)

# Invariant curve y=x
xinv = np.linspace(xlo,xhi,100)
yinv = [x for x in xinv]
plt.plot(xinv,yinv,c='k',lw=3)

# Vector field
plot_vector_field(lims,V)

# Formatting
plt.xlim(lims[0])
plt.ylim(lims[1])
```
