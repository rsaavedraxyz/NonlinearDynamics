# **Classifying Hopf bifurcations**

Consider the system of equations

$$
\left\lbrace
\begin{aligned}
&\dot{x} = y + \mu x \\
&\dot{y} = -x +\mu y - x^2y
\end{aligned}
\right.
$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_vector_field(_ax, _lims, _vfield, *kwargs):
    """
    Plot the vector field _vfield in a grid with limits _lims
    """
    xlims = _lims[0]
    ylims = _lims[1]
    xlo,xhi,ylo,yhi = xlims[0],xlims[1],ylims[0],ylims[1]
    XX, YY = np.meshgrid(np.arange(xlo,xhi,0.01), np.arange(ylo,yhi,0.01))
    DX, DY = _vfield([XX, YY], 0, *kwargs)
    _ax.streamplot(XX,YY,DX,DY,color='darkcyan',density=1.,minlength=0.2,linewidth=1,arrowsize=1.5)
    
def rk4(dxdt, x, t, dt, *args, **kwargs):
    x = np.asarray(x)
    k1 = np.asarray(dxdt(x, t, *args, **kwargs))*dt
    k2 = np.asarray(dxdt(x + k1*0.5, t, *args, **kwargs))*dt
    k3 = np.asarray(dxdt(x + k2*0.5, t, *args, **kwargs))*dt
    k4 = np.asarray(dxdt(x + k3, t, *args, **kwargs))*dt
    return x + (k1 + 2*k2 + 2*k3 + k4)/6
```

```python
def V(z, t, mu):
    x = z[0]
    y = z[1]
    dxdt = y + mu*x
    dydt = -x + mu*y - x**2*y
    return [dxdt, dydt]
```

```python
fig, ax = plt.subplots(1,2, sharex=True, figsize=(10, 5))

# Set boundaries
scl = 1.5
xlo,xhi,ylo,yhi = -1.*scl, 1.*scl, -1.*scl, 1.*scl
lims = [[xlo,xhi],[ylo,yhi]]

# Vector field
plot_vector_field(ax[0],lims,V,-0.1)
plot_vector_field(ax[1],lims,V,0.1)

# Formatting
plt.xlim(lims[0])
plt.ylim(lims[1])
```
