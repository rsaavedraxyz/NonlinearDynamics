# **Hopf bifurcation in a predator-prey model**

Consider the predator-prey model

$$
\left\lbrace
\begin{aligned}
&\dot{x} = x\left(b - x - \frac{y}{1+x}\right) \\
&\dot{y} = y\left(\frac{x}{1+x} - ay\right)
\end{aligned}
\right.
$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_vector_field(_lims, _vfield, *kwargs):
    """
    Plot the vector field _vfield in a grid with limits _lims
    """
    xlims = _lims[0]
    ylims = _lims[1]
    xlo,xhi,ylo,yhi = xlims[0],xlims[1],ylims[0],ylims[1]
    XX, YY = np.meshgrid(np.arange(xlo,xhi,0.01), np.arange(ylo,yhi,0.01))
    DX, DY = _vfield([XX, YY], 0, *kwargs)
    plt.streamplot(XX,YY,DX,DY,color='darkcyan',density=2.,minlength=0.2,linewidth=1,arrowsize=1.5)
```

```python
def V(z, t, a, b):
    x = z[0]
    y = z[1]
    dxdt = x*b - x**2 - x*y/(1+x)
    dydt = x*y/(1+x) - a*y**2
    return [dxdt, dydt]
```

```python


# Vector field and null-clines
xn = np.linspace(xlo,xhi,100)

fig = plt.subplots(sharex=True, figsize=(5, 5))
xlo,xhi,ylo,yhi = 1., 1.5, 3.5, 4.5
lims = [[xlo,xhi],[ylo,yhi]]
b = 3.
a = 4*(b-2)/b**2/(b+2) + 0.05
print("a=",a)
n1 = [-x**2 + (b - 1)*x + b for x in xn]
n2 = [x/a/(1+x) for x in xn]
plot_vector_field(lims,V,a,b)
plt.plot(xn,n1,c='r',ls='--')
plt.plot(xn,n2,c='r',ls='--')
plt.xlim(lims[0])
plt.ylim(lims[1])
#plt.set(xlim=lims[0], ylim=lims[1])

fig = plt.subplots(sharex=True, figsize=(5, 5))
xlo,xhi,ylo,yhi = 0.05, 0.25, 2.5, 4.
lims = [[xlo,xhi],[ylo,yhi]]
b = 3.
a = 4*(b-2)/b**2/(b+2) - 0.05
print("a=",a)
n1 = [-x**2 + (b - 1)*x + b for x in xn]
n2 = [x/a/(1+x) for x in xn]
plot_vector_field(lims,V,a,b)
plt.plot(xn,n1,c='r',ls='--')
plt.plot(xn,n2,c='r',ls='--')
plt.xlim(lims[0])
plt.ylim(lims[1])
#plt.set(xlim=lims[0], ylim=lims[1])

```
